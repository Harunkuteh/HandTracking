```python
import cv2
from cvzone.HandTrackingModule import HandDetector

# ‡πÄ‡∏õ‡∏¥‡∏î‡∏Å‡∏•‡πâ‡∏≠‡∏á
cap = cv2.VideoCapture(0)
detector = HandDetector(detectionCon=0.8)

# ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏ô‡∏¥‡πâ‡∏ß‡∏°‡∏∑‡∏≠
fingerNames = ["Thumb", "Index", "Middle", "Ring", "Pinky"]

# ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡∏´‡∏ô‡πâ‡∏≤‡∏ï‡πà‡∏≤‡∏á‡πÄ‡∏õ‡πá‡∏ô Fullscreen
cv2.namedWindow("Hand Tracking", cv2.WND_PROP_FULLSCREEN)
cv2.setWindowProperty("Hand Tracking", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    hands, frame = detector.findHands(frame)  # ‡∏ï‡∏£‡∏ß‡∏à‡∏à‡∏±‡∏ö‡∏°‡∏∑‡∏≠

    if hands:
        y_position = 50  # ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°
        for hand in hands:
            handType = hand["type"]  # ‡∏ã‡πâ‡∏≤‡∏¢‡∏´‡∏£‡∏∑‡∏≠‡∏Ç‡∏ß‡∏≤
            fingers = detector.fingersUp(hand)  # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ô‡∏¥‡πâ‡∏ß‡∏ó‡∏µ‡πà‡∏¢‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô
            fingerCount = fingers.count(1)  # ‡∏ô‡∏±‡∏ö‡∏ô‡∏¥‡πâ‡∏ß‡∏ó‡∏µ‡πà‡∏¢‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô

            # ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏ô‡∏¥‡πâ‡∏ß‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏¢‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô
            raisedFingers = [fingerNames[i] for i in range(5) if fingers[i] == 1]
            raisedFingersText = ", ".join(raisedFingers) if raisedFingers else "None"

            # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏™‡∏µ‡πÉ‡∏´‡πâ‡∏°‡∏∑‡∏≠‡∏ã‡πâ‡∏≤‡∏¢‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß ‡∏°‡∏∑‡∏≠‡∏Ç‡∏ß‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏ô‡πâ‡∏≥‡πÄ‡∏á‡∏¥‡∏ô
            color = (0, 255, 0) if handType == "Left" else (255, 0, 0)

            # ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
            cv2.putText(frame, f'{handType} Hand - Fingers: {fingerCount}', (20, y_position),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, color, 3, cv2.LINE_AA)
            y_position += 40
            cv2.putText(frame, f'Raised: {raisedFingersText}', (20, y_position),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 3, cv2.LINE_AA)
            y_position += 50  # ‡∏Ç‡∏¢‡∏±‡∏ö‡∏•‡∏á‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÑ‡∏°‡πà‡πÉ‡∏´‡πâ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ó‡∏±‡∏ö‡∏Å‡∏±‡∏ô

    # ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏ß‡∏¥‡∏î‡∏µ‡πÇ‡∏≠
    cv2.imshow("Hand Tracking", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):  # ‡∏Å‡∏î 'q' ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏¥‡∏î‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°
        break

cap.release()
cv2.destroyAllWindows()
```

# Modeling

### Core Implementation:
The model detects hands, counts raised fingers, and distinguishes between left and right hands.

### Features:

- üîÑ Mirror Effect: Flips the video frame for a natural mirror view.
- üìù Dynamic Text Display: Shows hand type, finger count, and raised finger names in real time.

## Main Code

```python
import cv2
from cvzone.HandTrackingModule import HandDetector

# Open the camera
cap = cv2.VideoCapture(0)
detector = HandDetector(detectionCon=0.9, maxHands=2)

# List of finger names
fingerNames = ["Thumb", "Index", "Middle", "Ring", "Pinky"]

# Set the window to Fullscreen
cv2.namedWindow("Hand Tracking", cv2.WND_PROP_FULLSCREEN)
cv2.setWindowProperty("Hand Tracking", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)  # üîÑ Mirror the frame
    hands, frame = detector.findHands(frame)  # ü§ñ Detect hands

    if hands:
        y_position = 30  # Starting vertical position for text
        for hand in hands:
            handType = hand["type"]  # Left or Right
            fingers = detector.fingersUp(hand)  # Check raised fingers
            fingerCount = fingers.count(1)  # Count raised fingers

            # List of raised fingers
            raisedFingers = [fingerNames[i] for i in range(5) if fingers[i] == 1]
            raisedFingersText = ", ".join(raisedFingers) if raisedFingers else "None"

            # Set text color: green for left hand, blue for right hand
            color = (0, 255, 0) if handType == "Left" else (255, 0, 0)

            # Display hand type and finger count
            cv2.putText(frame, f'{handType} Hand - {fingerCount} fingers', (20, y_position),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2, cv2.LINE_AA)
            y_position += 25
            # Display which fingers are raised
            cv2.putText(frame, f'Raised: {raisedFingersText}', (20, y_position),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2, cv2.LINE_AA)
            y_position += 30  # Prevent text overlap

    # Show the video
    cv2.imshow("Hand Tracking", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from cvzone.HandTrackingModule import HandDetector
import matplotlib.pyplot as plt

# ‡πÄ‡∏õ‡∏¥‡∏î‡∏Å‡∏•‡πâ‡∏≠‡∏á
cap = cv2.VideoCapture(0)
detector = HandDetector(detectionCon=0.95, maxHands=2)  # ‡πÄ‡∏û‡∏¥‡πà‡∏° detectionCon ‡πÉ‡∏´‡πâ‡∏™‡∏π‡∏á‡∏Ç‡∏∂‡πâ‡∏ô

# ‡∏ä‡∏∑‡πà‡∏≠‡∏Ç‡∏≠‡∏á‡∏ô‡∏¥‡πâ‡∏ß
fingerNames = ["Thumb", "Index", "Middle", "Ring", "Pinky"]

# ‡∏ï‡∏±‡∏ß‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà‡∏ô‡∏¥‡πâ‡∏ß‡∏ñ‡∏π‡∏Å‡∏¢‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô
fingerCountDict = {finger: 0 for finger in fingerNames}

cv2.namedWindow("Hand Tracking", cv2.WND_PROP_FULLSCREEN)
cv2.setWindowProperty("Hand Tracking", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
previous_fingers = [0] * 5  # ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Å‡∏≤‡∏£‡∏¢‡∏Å‡∏ô‡∏¥‡πâ‡∏ß‡πÉ‡∏ô‡πÄ‡∏ü‡∏£‡∏°‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤
correct_predictions = 0
total_frames = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    hands, frame = detector.findHands(frame)

    if hands:
        y_position = 30
        for hand in hands:
            handType = hand["type"]
            fingers = detector.fingersUp(hand)

            # ‡∏≠‡∏±‡∏õ‡πÄ‡∏î‡∏ï‡∏ï‡∏±‡∏ß‡∏ô‡∏±‡∏ö‡∏Ç‡∏≠‡∏á‡∏ô‡∏¥‡πâ‡∏ß‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏¢‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô
            for i in range(5):
                if fingers[i] == 1:
                    fingerCountDict[fingerNames[i]] += 1

            # ‡∏ï‡∏£‡∏ß‡∏à‡∏à‡∏±‡∏ö‡∏ô‡∏¥‡πâ‡∏ß‡∏ó‡∏µ‡πà‡∏¢‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô‡πÅ‡∏•‡∏∞‡∏Å‡∏£‡∏≠‡∏á‡∏ú‡∏•‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ
            raisedFingers = [fingerNames[i] for i in range(5) if fingers[i] == 1]
            correct_predictions += sum([previous_fingers[i] == 1 and fingers[i] == 1 for i in range(5)])

            # ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô‡πÄ‡∏ü‡∏£‡∏°‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô
            previous_fingers = fingers

            # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏™‡∏µ‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°
            color = (0, 255, 0) if handType == "Left" else (255, 0, 0)

            # ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ö‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡∏à‡∏≠
            cv2.putText(frame, f'{handType} Hand - {fingers.count(1)} fingers', 
                        (20, y_position), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
            y_position += 30

            # ‡πÅ‡∏™‡∏î‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏Ç‡∏≠‡∏á‡∏ô‡∏¥‡πâ‡∏ß‡∏ó‡∏µ‡πà‡∏¢‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô
            raisedFingersText = ", ".join(raisedFingers) if raisedFingers else "None"
            cv2.putText(frame, f'Raised: {raisedFingersText}', 
                        (20, y_position), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
            y_position += 30  # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏£‡∏∞‡∏¢‡∏∞‡∏´‡πà‡∏≤‡∏á‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î

    # ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏†‡∏≤‡∏û
    cv2.imshow("Hand Tracking", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    total_frames += 1

cap.release()
cv2.destroyAllWindows()

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Accuracy
accuracy = correct_predictions / total_frames * 100 if total_frames > 0 else 0
print(f"Accuracy: {accuracy:.2f}%")

# ‡πÅ‡∏™‡∏î‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡∏Ç‡∏≠‡∏á‡∏ô‡∏¥‡πâ‡∏ß‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏¢‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô
plt.figure(figsize=(8, 5))
plt.bar(fingerCountDict.keys(), fingerCountDict.values(), color=['red', 'blue', 'green', 'purple', 'orange'])
plt.xlabel("Finger")
plt.ylabel("Count")
plt.title("Finger Raise Frequency")
plt.xticks(rotation=45)  # ‡∏´‡∏°‡∏∏‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡∏Ç‡∏≠‡∏á‡∏ô‡∏¥‡πâ‡∏ß‡πÉ‡∏´‡πâ‡∏ä‡∏±‡∏î‡πÄ‡∏à‡∏ô
plt.show()
```
